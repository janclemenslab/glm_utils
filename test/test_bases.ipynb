```python
import numpy as np
import matplotlib.pyplot as plt
from glm_utils.bases import bsplines, laplacian_pyramid, raised_cosine
```

```python
B = laplacian_pyramid(width=100, levels=4, step=1, FWHM=10)
plt.gcf().set_size_inches(10, 10)
plt.subplot(211)
plt.plot(B);
plt.xlabel('time')
plt.subplot(212)
plt.imshow(B);
plt.ylabel('time')
plt.xlabel('basis functions')
B.shape
```

```python
B = bsplines(width=100, positions=np.arange(5,100,5), periodic=False)
plt.gcf().set_size_inches(10, 10)
plt.subplot(211)
plt.plot(B.T);
plt.subplot(212)
plt.imshow(B);
B.shape
```

```python
B = raised_cosine(neye=0, ncos=10, kpeaks=(5, 70), b=2, nkt=10)
plt.gcf().set_size_inches(10, 10)
plt.subplot(211)
plt.plot(B.T);
plt.subplot(212)
plt.imshow(B);
B.shape
```
